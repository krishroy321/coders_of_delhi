```python
import pandas as pd
```

```python
employees = pd.DataFrame({
    "Emp ID":[1,2,3],
    "Name" : ["Alice","Bob","Charlie"],
    "DeptID" : [10,20,30]
})

departments = pd.DataFrame({
    "DeptID" : [10,20,40],
    "DeptName" : ["HR","Engineering","Marketing"]
})
```

```python
employees
```

```python
departments
```

```python
pd.merge(employees,departments,on="DeptID")
```

```python
pd.merge(employees,departments,on="DeptID",how="left")
```

```python
pd.merge(employees,departments,on="DeptID",how="right")
```

```python
pd.merge(employees,departments,on="DeptID",how="outer")
```

CONCATENATING DATAFRAMES

```python
df1 = pd.DataFrame({"Name":["Alice","Bob"],"Score":[423,345]})
df2 = pd.DataFrame({"Name":["Charlie","David"],"Age":[23,45]})
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],axis=1)
```
