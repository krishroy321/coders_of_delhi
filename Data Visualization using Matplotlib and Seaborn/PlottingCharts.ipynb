```python
years = [1990,1992,1994,1996,1998,2000,2003,2005,2007,2010]
runs = [500,700,1100,1500,1800,1200,1700,1300,900,1500]
```

```python
plt.plot(years,runs)
```

```python
plt.plot(years,runs)
plt.xlabel('Years')
plt.ylabel('Runs scored')
plt.title("Sachin Tendulkar's Yearly Runs")
```

```python
kohli = [0,0,500,800,1100,1300,1500,1800,1900,2100]
sehwag = [0,300,800,1200,1500,1700,1600,1400,1000,0]

plt.plot(years,kohli , label="Virat Kohli")
plt.plot(years,sehwag,label="Virender Sehwag")

plt.xlabel("Years")
plt.ylabel("Runs")
plt.title("Performance Comparison")
plt.legend()
plt.show()
```

```python
plt.plot(years,kohli, 'ro--',label="Virat Kohli")
plt.plot(years,sehwag,'g^:',label="Virender Sehwag")
plt.legend()
```

```python
plt.plot(years,kohli,color="Purple",linestyle='--',label="kohli")
plt.plot(years,sehwag,color="Red",linestyle='-.',label="sehwag")
plt.plot(years,runs,color="blue",label="Sachin")
plt.tight_layout()
plt.grid(True)
plt.legend()
```
