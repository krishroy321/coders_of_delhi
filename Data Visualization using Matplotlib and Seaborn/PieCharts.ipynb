```python
import matplotlib.pyplot as plt

plt.style.use("ggplot")
```

```python
players = ["Sachin","Sehwag","Kohli","Yuvraj"]
runs = [18000,8000,12000,9500]

plt.title("Carrer runs of the Indian Batsmen")
plt.pie(runs,labels=players)
plt.show()
```

```python
colors=["#6A99EB","#98DE5B","#1B9BD1","#E82542"]
plt.pie(runs,labels=players,colors=colors)
plt.show()
```

```python
plt.pie(
    runs,
    labels=players,
    colors=colors,
    wedgeprops={'edgecolor':'black','linewidth':2,'linestyle':'-.','fill':True}
)
plt.show()
```

```python
explode=[0.1,0,0,0]
plt.pie(runs,labels=players,colors=colors,explode=explode,shadow=True,startangle=90,autopct='%1.1f%%')
plt.show()
```
